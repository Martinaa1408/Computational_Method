1. Distanza di Levenshtein

La Distanza di Levenshtein è una misura della differenza tra due stringhe.

Sostituzioni (X) → Cambiare un carattere in un altro.

Inserzioni (^) → Aggiungere un carattere.

Cancellazioni (v) → Rimuovere un carattere.

📌 Regola importante:
Se due sequenze hanno la stessa lunghezza, la Distanza di Hamming è sempre maggiore o uguale alla Distanza di Levenshtein.

Distanza di Levenshtein è più flessibile, perché permette di allineare le sequenze e ottenere una distanza più accurata.

L'algoritmo di Needleman-Wunsch trova il miglior allineamento globale tra due sequenze.

Un allineamento locale è un allineamento tra sottosequenze delle sequenze target.

L'algoritmo di Smith-Waterman è una modifica dell'algoritmo di NW per trovare gli allineamenti locali ottimali.

- Nessun gap iniziale o finale nelle sequenze allineate.

- Scarta le parti dell'allineamento che avrebbero un punteggio negativo per mantenere solo le regioni con somiglianza significativa.

- L'allineamento inizia dalla cella con il punteggio più alto (e non in basso a destra come in Needleman-Wunsch).
