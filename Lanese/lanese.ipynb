CICLO FOR e CICLO WHILE

FOR-->Iterare su una sequenza predefinita (es. lista, range, stringa).

WHILE-->Eseguire un blocco di codice finché una condizione è vera.

```python
#for variabile in sequenza:
#blocco di codice

numeri = [1, 2, 3, 4]
for numero in numeri:
    print(numero)
# Output:
# 1
# 2
# 3
# 4

for i in range(5):  # Itera da 0 a 4
    print(i)
# Output:
# 0
# 1
# 2
# 3
# 4
```

```python
#while condizione:
# blocco di codice

x = 0
while x < 5:
    print(x)
    x += 1
# Output:
# 0
# 1
# 2
# 3
# 4

while True:
    risposta = input("Vuoi continuare? (s/n): ")
    if risposta == "n":
        break  # Esce dal ciclo
```

CASI IN CUI SI POSSONO USARE ENTRAMBI

```python
#STAMPARE I NUMERI DA 1 A 5
for i in range(1, 6):
    print(i)

i = 1
while i <= 5:
    print(i)
    i += 1

#SOMMARI I NUMERI IN UNA LISTA
numeri = [1, 2, 3, 4, 5]
somma = 0
for numero in numeri:
    somma += numero
print(somma)  # Output: 15

numeri = [1, 2, 3, 4, 5]
somma = 0
i = 0
while i < len(numeri):
    somma += numeri[i]
    i += 1
print(somma)  # Output: 15

```

Quando usare FOR:
Quando conosci il numero di iterazioni.
Quando stai iterando su una sequenza (lista, stringa, range, dizionario).

Quando usare WHILE:
Quando non conosci il numero di iterazioni in anticipo.
Quando il ciclo dipende da una condizione.

```python
for i in range(10):
    if i == 5:
        break  # Esce dal ciclo
    print(i)
# Output: 0 1 2 3 4

for i in range(10):
    if i % 2 == 0:
        continue  # Salta i numeri pari
    print(i)
# Output: 1 3 5 7 9
```

CONDIZIONALI

I condizionali vengono utilizzati per eseguire blocchi di codice solo se una determinata condizione è vera.



```python
#SINTASSI DI BASE
if condizione:
# Blocco di codice se condizione è vera
elif altra_condizione:
# Blocco di codice se altra_condizione è vera
else:


# Blocco di codice se nessuna condizione è vera


def calcolatrice(a, b, operazione):
    if operazione == "somma":
        return a + b
    elif operazione == "sottrazione":
        return a - b
    elif operazione == "moltiplicazione":
        return a * b
    elif operazione == "divisione":
        if b != 0:
            return a / b
        else:
            return "Errore: divisione per zero"
    else:
        return "Operazione non valida"


print(calcolatrice(10, 5, "somma"))  # Output: 15
print(calcolatrice(10, 5, "divisione"))  # Output: 2.0
print(calcolatrice(10, 0, "divisione"))  # Output: Errore: divisione per zero

```

RECURSIVE FUNCTION


```python
def fattoriale(n):
    if n == 0:  # Caso base
        return 1
    else:
        return n * fattoriale(n - 1)  # Richiamo ricorsivo


# Test
print(fattoriale(5))  # Output: 120
```

FIBONACCI

```python
def fibonacci(n):
    if n <= 1:  # Caso base
        return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)  # Richiamo ricorsivo


# Test
print(fibonacci(6))  # Output: 8
```

SOMMA LISTA

```python
def somma_lista(lista):
    if not lista:  # Caso base: lista vuota
        return 0
    else:
        return lista[0] + somma_lista(lista[1:])  # Richiamo ricorsivo


# Test
print(somma_lista([1, 2, 3, 4]))  # Output: 10
```

REGULAR EXPRESSION o regexp sono una potente sintassi per cercare, confrontare e manipolare stringhe. Python fornisce il modulo re per lavorare con le regexp.


Usa r"..." per creare una stringa raw ed evitare problemi con i caratteri di escape.

Usa re.match per verificare l'inizio della stringa.

Usa re.search per cercare un pattern ovunque nella stringa.

Usa re.findall per ottenere tutte le occorrenze del pattern.

Usa re.sub per sostituire i pattern trovati con un'altra stringa.

```python
import re

telefono = "+39 333-123-4567"
pattern = r"^\+?\d{1,4}\s\d{3}-\d{3}-\d{4}$"
if re.match(pattern, telefono):
    print("Numero di telefono valido!")
else:
    print("Numero non valido.")

import re

data = "27/01/2025"
pattern = r"^(0[1-9]|[12][0-9]|3[01])/(0[1-9]|1[0-2])/\d{4}$"
if re.match(pattern, data):
    print("Data valida!")
else:
    print("Data non valida.")

import re


def verifica_dna(sequenza):
    pattern = r"^[ATCG]+$"  # Solo caratteri A, T, C, G
    if re.match(pattern, sequenza):
        return "Sequenza valida"
    else:
        return "Sequenza non valida"


# Test
print(verifica_dna("ATCGATCG"))  # Output: Sequenza valida
print(verifica_dna("ATXG"))  # Output: Sequenza non valida
```

CLASSI

è una struttura che permette di creare oggetti, con attributi (proprietà) e metodi (funzioni).


```python
#CASO BASE
class Persona:
    def __init__(self, nome, età):  # Costruttore
        self.nome = nome  # Attributo di istanza
        self.età = età

    def saluta(self):  # Metodo
        print(f"Ciao, mi chiamo {self.nome} e ho {self.età} anni.")


# Creazione di un oggetto
mario = Persona("Mario", 30)
mario.saluta()  # Output: Ciao, mi chiamo Mario e ho 30 anni.
```

ATTRIBUTI E METODI DI CLASSE


```python
class Persona:
    specie = "Homo sapiens"  # Attributo di classe

    def __init__(self, nome, età):
        self.nome = nome  # Attributo di istanza
        self.età = età

    @classmethod
    def descrizione_specie(cls):  # Metodo di classe
        print(f"Siamo tutti {cls.specie}.")


# Creazione di oggetti
persona1 = Persona("Alice", 25)
persona2 = Persona("Luigi", 35)

# Accesso agli attributi di classe e metodo di classe
print(persona1.specie)  # Output: Homo sapiens
Persona.descrizione_specie()  # Output: Siamo tutti Homo sapiens.
```

L'ereditarietà permette di creare una nuova classe basata su una classe esistente.

```python
class Persona:
    def __init__(self, nome, età):
        self.nome = nome
        self.età = età

    def saluta(self):
        print(f"Ciao, mi chiamo {self.nome} e ho {self.età} anni.")


class Studente(Persona):  # Studente eredita da Persona
    def __init__(self, nome, età, corso):
        super().__init__(nome, età)  # Richiama il costruttore della classe base
        self.corso = corso

    def info_studente(self):
        print(f"Sono uno studente di {self.corso}.")


# Creazione di un oggetto della sottoclasse
studente = Studente("Maria", 20, "Informatica")
studente.saluta()  # Output: Ciao, mi chiamo Maria e ho 20 anni.
studente.info_studente()  # Output: Sono uno studente di Informatica.
```

Concetto di "palindrome complementare"

Complemento: In DNA, ogni base ha una base complementare:
A ↔ T
C ↔ G

Palindromo: Una sequenza palindromica è identica al suo complemento inverso.
Ad esempio, ATGCAT è palindromica perché il complemento inverso è ancora ATGCAT.

```python
seq1 = input("Insert a DNA sequence:")
seq2 = input("Insert a 2nd DNA sequence:")
bases = "ACGT"
diff_seq = ""  #Questa stringa conterrà i simboli di confronto (| o X).
for i in range(len(seq1)):
    pos1 = bases.find(seq1[i])  #indice pos1
    pos2 = bases.find(seq2[i])
    if pos1 + pos2 == 3:  #Se la somma degli indici (pos1 + pos2) è 3, viene aggiunto un | a diff_seq. Questo rappresenta una coppia complementare valida:
        diff_seq = diff_seq + "|"
    else:
        diff_seq = diff_seq + "X"
print(seq1)
print(diff_seq)
print(seq2)

seq1 = input("Enter the first DNA sequence: ").upper()
seq2 = input("Enter the second DNA sequence: ").upper()
if len(seq1) != len(seq2):
    print("Sequences must be of the same length!")
else:
    middle_line = ""
    non_complementary_count = 0
    for i in range(len(seq1)):
        base1 = seq1[i]
        base2 = seq2[i]
        if (base1 == 'A' and base2 == 'T') or (base1 == 'T' and base2 == 'A') or \
                (base1 == 'C' and base2 == 'G') or (base1 == 'G' and base2 == 'C'):
            middle_line += '|'
        else:
            middle_line += 'X'
            non_complementary_count += 1
    print(seq1)
    print(middle_line)
    print(seq2)
    print(non_complementary_count)
```

```python
#B: write a python program from file 'guess.txt' a single line containing space-separated integers. The user plays the following game:he insert a number and needs to guess the sum of two numbers in the line. If the guess is correct he is congratulated with 'you win!' and the game ends. if the guess is wrong he has to retry but gets: 'upper!' if all the possible sums are higher than the guessed numbers, 'lower!' if they are all lower and 'try again!' otherwise.
#Input:  4 12 13 5 
#run:
#8 Upper!
#22 Try again!
#25 you win!
def guessed_sum(input_file):
    # Legge i numeri dal file
    with open(input_file, "r") as reader:
        line = reader.readline().strip()
        elements = line.split()
        numbers = []

        for num in elements:
            numbers.append(int(num))  # Converte ogni numero in intero e lo aggiunge alla lista

    # Genera tutte le possibili somme di due numeri diversi
    possible_sums = []

    for i in range(len(numbers)):
        for j in range(i + 1, len(numbers)):  # Evita somme dello stesso numero
            possible_sums.append(numbers[i] + numbers[j])

    # Ciclo di gioco senza limite di tentativi
    while True:
        guess = int(input("Enter your guess: "))

        if guess in possible_sums:
            print("You win!")
            break  # Termina il gioco

        all_higher = True
        all_lower = True

        for s in possible_sums:
            if guess < s:
                all_lower = False
            if guess > s:
                all_higher = False

        if all_higher:
            print("Lower!")  # Se tutti i numeri sono più piccoli del guess
        elif all_lower:
            print("Upper!")  # Se tutti i numeri sono più grandi del guess
        else:
            print("Try again!")  # Se il numero è tra i valori ma non è corretto


# Avvia il gioco
guessed_sum("guess.txt")

```

```python
#Write a python program to play the hangman game.
#It reads words from a file (whose name is taken as command line parameter), and plays rounds on the hangman game using them till the user asks to stop.
#The game works as follows. The schema of the word obtained by replacing vowels with +, consonants with – and any other symbol with ? is printed on the screen. The user can insert a letter, or guess the word.
#In the first case, all occurrences of the letter in the word are shown, and the game continues.
#In the second case, if the guess is correct then the user wins, and the game shows the number of total tries, otherwise the game continues.

#Ex. (assuming the first word in the file is CLASSROOM, user input in red)
#--+---++-
#Insert a letter or guess the word:A
#--A---++-
#Insert a letter or guess the word:E
#--A---++-
#Insert a letter or guess the word:O
#--A---OO-
#Insert a letter or guess the word:S
#--ASS-OO-
#Insert a letter or guess the word:T
#--ASS-OO-
#Insert a letter or guess the word:GLASSNOON
#Nice try! ... but wrong...
#--ASS-OO-
#Insert a letter or guess the word:CLASSROOM
#You win! In 7 tries!
#Do you want to play again Y/N? N

from sys import argv

# Apertura del file
file_word = open(argv[1], 'r')

# Dizionari per vocali e consonanti
dictionary_vowels = {
    'A': '+',
    'E': '+',
    'I': '+',
    'O': '+',
    'U': '+'
}

dictionary_consonants = {
    'B': '-',
    'C': '-',
    'D': '-',
    'F': '-',
    'G': '-',
    'H': '-',
    'J': '-',
    'K': '-',
    'L': '-',
    'M': '-',
    'N': '-',
    'P': '-',
    'Q': '-',
    'R': '-',
    'S': '-',
    'T': '-',
    'V': '-',
    'W': '-',
    'X': '-',
    'Y': '-',
    'Z': '-'
}


# Funzione per creare lo schema della parola
def create_schema(
        word):  #Le vocali vengono rappresentate come '+'. Le consonanti come '-'. Caratteri non riconosciuti (es. numeri o simboli) diventano ?.
    schema = ''
    for letter in word:
        if letter in dictionary_vowels:
            schema += dictionary_vowels[letter]
        elif letter in dictionary_consonants:
            schema += dictionary_consonants[letter]
        else:
            schema += '?'
    return schema


# Funzione per aggiornare lo stato della parola
def update_display(word,
                   guessed_letters):  #Aggiorna il progresso del giocatore: Le lettere indovinate (presenti in guessed_letters) vengono mostrate. Le lettere non indovinate rimangono come _.
    display = ''
    for letter in word:
        if letter in guessed_letters:  #Lista delle lettere già provate dal giocatore
            display += letter
        else:
            display += '_'
    return display


# Funzione principale del gioco
def play_game(word):
    guessed_letters = []
    tries = 0
    print("Ecco lo schema della parola:", create_schema(word))

    while True:
        display = update_display(word, guessed_letters)
        print("Parola attuale:", display)

        if '_' not in display:  # Se tutte le lettere sono state indovinate
            print(f"Hai indovinato la parola '{word}' in {tries} tentativi!")
            break

        user_input = input("Inserisci una lettera o prova a indovinare la parola: ").strip().upper()
        tries += 1

        if user_input == word:  # Se l'utente indovina la parola
            print(f"Complimenti! Hai indovinato la parola '{word}' in {tries} tentativi!")
            break
        elif len(user_input) == 1:  # Se l'utente inserisce una lettera
            if user_input in guessed_letters:
                print(f"Hai già provato la lettera '{user_input}'!")
            else:
                guessed_letters.append(user_input)
                print(f"La lettera '{user_input}' è {'corretta!' if user_input in word else 'sbagliata!'}")
        else:  # Se l'utente prova a indovinare ma sbaglia
            print("Tentativo errato. Continua a giocare!")


# Ciclo per ogni parola nel file
for line in file_word:
    word = line.strip().upper()
    play_game(word)
    play_again = input("Vuoi giocare di nuovo? (Y/N): ").strip().upper()
    if play_again != 'Y':
        print("Grazie per aver giocato. Alla prossima!")
        break

file_word.close()
#Il programma legge parole da un file.Per ogni parola:
#Mostra uno schema iniziale basato su vocali e consonanti.
#L'utente indovina lettere o l'intera parola.
#Tiene traccia dei tentativi e delle lettere usate.
#Il ciclo continua finché il filemanfredi non è esaurito o il giocatore decide di smettere.
```

```python
#Write a python program that takes a source file containing in each line an arithmetic integer 
#expression with only sums and subtractions, and computes the results. The name of the 
#source file is taken as command line parameter. If there is an additional command line 
#parameter then this is the name of the file where the results will be written, otherwise the 
#results are printed on the screen. 
#Input
#34+12-2
#10-23+1
#100+3+12+2
#2
#Output
#34+12-2=44
#10-23+1=-12
#100+3+12+2=117
#2=2   
def evaluate_expressions(input_filename, output_filename):
    file = open(input_filename, "r")
    lines = file.readlines()
    file.close()

    results = []

    for line in lines:
        line = line.strip()
        if line == "":
            continue

        numbers = []
        operators = []
        num = ""

        # Separare numeri e operatori in modo semplice
        for char in line:
            if char == "+" or char == "-":
                numbers.append(int(num))  # Salva il numero precedente
                operators.append(char)  # Salva l'operatore
                num = ""  # Resetta la stringa per il numero successivo
            else:
                num += char  # Costruisce il numero

        numbers.append(int(num))  # Aggiunge l'ultimo numero trovato

        # Calcolo dell'espressione manualmente
        total = numbers[0]
        i = 0
        while i < len(operators):
            if operators[i] == "+":
                total += numbers[i + 1]
            else:
                total -= numbers[i + 1]
            i += 1

        results.append(line + "=" + str(total))  # Memorizza il risultato

    # Stampa o scrive il risultato in un file
    if output_filename:
        file = open(output_filename, "w")
        for result in results:
            file.write(result + "\n")
        file.close()
    else:
        for result in results:
            print(result)


# Input utente per i file
input_filename = input("Enter input filename: ")
output_filename = input("Enter output filename (press Enter to print on screen): ")
if output_filename == "":
    output_filename = None

evaluate_expressions(input_filename, output_filename)
```

```python
#Write a python program to do spellchecking. It has two input files, text.txt and dict.txt, and 
#one output file, output.txt. File dict.txt contains a word per line, and represents the correct 
#words. File text.txt represents a text, written for simplicity with a word per line as well. File 
#output.txt is obtained from text.txt as follows. For each word, if the word is correct, it is 
#copied to output.txt. If it is not, the program asks to the user whether the word should be 
#considered correct. If so, it is written on output.txt, and from now on it becomes a correct 
#word. If not, it asks the correct version to the user, and the check for correctness restarts.

def spell_checker(text_file, dict_file, output_file):
    # Legge il dizionario
    file = open(dict_file, "r")
    dictionary = []
    for word in file:
        dictionary.append(word.strip())  # Rimuove spazi extra e va a capo
    file.close()

    # Legge il testo da controllare
    file = open(text_file, "r")
    words = []
    for word in file:
        words.append(word.strip())  # Rimuove spazi extra
    file.close()

    # Lista per l'output
    output = []

    # Controllo ortografico
    for word in words:
        if word in dictionary:
            output.append(word)
        else:
            print("Unknown word: " + word)
            correction = input("Is this word correct? (y/n) ")
            if correction.lower() == "y":
                dictionary.append(word)  # Aggiunge al dizionario
                output.append(word)
            else:
                correct_word = input("Enter the correct word: ")
                dictionary.append(correct_word)  # Aggiunge la parola corretta
                output.append(correct_word)

    # Scrive il risultato in output.txt
    file = open(output_file, "w")
    for word in output:
        file.write(word + "\n")
    file.close()


# Chiede i nomi dei file all'utente e avvia il correttore
text_file = input("Enter text filename: ")
dict_file = input("Enter dictionary filename: ")
output_file = input("Enter output filename: ")

spell_checker(text_file, dict_file, output_file)
```

```python
#Write a python program which takes from command line a file name and an optional -d flag.
#The file needs to be a file of space-separated integers. The program prints on the screen the 
#longest strictly increasing sub-sequence. If two strictly increasing sub-sequences have the 
#same length, it prints the first one. If the optional -d flag is present, it prints instead the 
#longest strictly decreasing sub-sequence.
#Example (without -d)
#23 10 13 17 14 18 20 9 11
#10 13 17 
from sys import argv  # Importa solo argv invece di tutto sys


def find_longest_subsequence(filename, decreasing):
    file = open(filename, "r")
    line = file.readline().strip()
    file.close()

    numbers = []
    for num in line.split():
        numbers.append(int(num))

    longest = []
    current = []

    for i in range(len(numbers)):
        if i == 0 or (numbers[i] > numbers[i - 1] and not decreasing) or (numbers[i] < numbers[i - 1] and decreasing):
            current.append(numbers[i])
        else:
            if len(current) > len(longest):
                longest = current
            current = [numbers[i]]

    if len(current) > len(longest):
        longest = current

    result = ""
    for num in longest:
        result += str(num) + " "

    print(result.strip())


# Usa argv senza sys.
if len(argv) < 2:
    print("Usage: python script.py <filename> [-d]")
else:
    filename = argv[1]
    flag = "-d" in argv  # Controlla se "-d" è presente negli argomenti
    find_longest_subsequence(filename, flag)
```

```python
#Write a Python program that takes from the user strings containing comma-separated integers. The program takes as command line parameters an integer threshold and the name of a destination file to write the results.
#For each line, the program prints on the destination file the number of pairs of integers in the line whose product is strictly greater than the threshold.

def count_pairs_above_threshold(numbers, threshold):
    """Conta le coppie di numeri il cui prodotto è maggiore della soglia."""
    count = 0

    for i in range(len(numbers)):
        for j in range(i + 1, len(numbers)):
            if numbers[i] * numbers[j] > threshold:
                count += 1

    return count


def process_file(input_file, output_file, threshold):
    """Legge un file, conta le coppie per ogni riga e scrive i risultati in un altro file."""

    with open(input_file) as reader, open(output_file, 'w') as writer:

        for line in reader:
            try:
                numbers = []

                for x in line.strip().split(","):
                    numbers.append(int(x))  # Converte gli elementi in interi

                result = count_pairs_above_threshold(numbers, threshold)

                writer.write(str(result) + "\n")

            except ValueError:
                print("Errore: riga non valida '" + line.strip() + "' ignorata.")


# Esegui il programma con try-except solo alla fine
try:
    process_file("line.txt", "output.txt", 18)

    print("Elaborazione completata. Risultati salvati in 'output.txt'.")

except FileNotFoundError:
    print("Errore: il file 'line.txt' non esiste.")

except Exception as e:
    print("Errore imprevisto:", e)
```
