# **Differenza tra linguaggi di programmazione compilati ed interpretati**
## Linguaggi di programmazione compilati
I linguaggi di programmazione compilati come il C, il C++, Visual Basic, ecc...seguono i seguenti passaggi:
- scrittura del codice in un editor
- compilazione del codice: il codice viene trasformato nel corrispondente codice in linguaggio macchina, generando un file eseguibile
- esecuzione del codice da parte del processore.
Il vantaggio di utilizzare linguaggi di programmazione compilati è che risultano più veloci nelle prestazioni; di contro però la sintassi utilizzata da questi linguaggi di programmazione è più complicata e macchinosa.
## Linguaggi di programmazione interpretati
Nei linguaggi di programmazione interpretati come ad esempio PHP, Python, ecc...il codice sorgente viene interpretato e se sintatticamente corretto, vengono eseguite le istruzioni così come descritte nel codice sorgente, quindi senza la generazione di un file eseguibile.
Il vantaggio di questi linguaggi di programmazione è l'elevata portabilità e immediatezza tra quello che scriviamo e quello che viene eseguito, di contro però questi linguaggi comportano un maggior carico di lavoro per il processore.

# **Differenza tra programmazione procedurale e programmazione ad oggetti**
## Programmazione procedurale
Nella programmazione procedurale il concetto base è quello di raggruppare i pezzi di programma ripetuti in porzioni di codice utilizzabili e richiamabili ogni volta che se ne presenti l’esigenza. Ogni procedura può essere vista come un sottoprogramma che svolge una ben determinata funzione (ad esempio, il calcolo della radice quadrata di un numero) e che è visibile e richiamabile dal resto del codice.
## Programmazione ad oggetti
La programmazione ad oggetti si basa sul fatto che esistono una serie di oggetti che interagiscono vicendevolmente, scambiandosi messaggi ma mantenendo ognuno il proprio stato ed i propri dati. 
Dunque mentre nella programmazione procedurale tradizionale devo scrivere tutte le assegnazioni per ogni oggetto (dato). Nella programmazione a oggetti, invece, mi basta assegnare le proprietà a una classe. Poi associo ogni oggetto alla classe e l'oggetto eredita automaticamente tutte le proprietà della classe.

## **HIGH ORDER FUNCTIONS**

Higher-order functions sono funzioni che utilizzano come parametro un'altra funzione, semplifcandone l'applicazione ma complicandone la scrittura: si avranno più funzioni definite che possono essere richiamate da altre funzioni. 

```python
#Funzione per trovare i mismatches
def find_mismatch(nucl1,nucl2):
    bases = "ACGT"
    pos1 = bases.find(nucl1)
    pos2 = bases.find(nucl2)
    return pos1 + pos2 == 3

def equal(x,y):
    return x==y

#in questo caso utilizzo un terzo parametro che in verità è una funzione, questa
#all'intero della funzione andrà a definire il parametro di complementare: per esempio
#equal se le sequenze sono uguali, ma può anche essere un altro

def find_mismatches(se1,seq2,comp):
    diff_seq = ""
    for i in range(len(seq1)):
    if comp(seq1[i],seq2[i]): # 
        diff_seq = diff_seq + "|"
    else:
        diff_seq = diff_seq + "X"
        return(diff_seq)

```

# **CICLO WHILE**

***

Il while è un iteratore che consente di eseguire le istruzioni presenti al suo interno finché la condizione restituisce come valore True. Quando la condizione non sarà più True ma False, il programma uscirà dal loop.

Se la condizione di controllo restasse perennemente True allora il programma entrerebbe in un loop infinito. Ad esempio utilizzando `while True:` si entra in un ciclo infinito. il ciclo while risulta essere più generale del ciclo for, per questo viene utilizzato ad esempio se non so quante volte devo ripetere il loop in questione.
 
Si possono usare variabili operatori etc etc per definire la condizione del loop. 

```python
# Struttura
while(condizione):
    |
    istruzioni
    |
    
```

```python
while(a>1)
while(True)
```

## **Istruzione break**

Tale istruzione, che si trova nel blocco delle istruzioni dell'iteratore, permette di interrompere un ciclo prematuramente e saltarne fuori.
In alcuni programmi se stai definendo una funzione, *return* può avere lo stesso effetto. Utilizzando come condizione **True**, il loop continua ad andare avanti all'infinito, fino a quando non aggiungo una funzione **Break** o **Return** per stopparlo.

```python
# Esempio:
while True:
    if i == 3:
        break
```

## **Istruzione continue**

Tale istruzione fa saltare tutto il codice dopo l'istruzione e fa ripartire Python dalla prima riga del ciclo

```python
# Esempio:
skip = 5
counter = 0

while counter < 10:
    counter += 1
    if counter == skip:
        print("Sto saltando " + str(skip))
        continue 
    print(counter)
# Senza il continue verrebbe stampato anche il numero 5
```

# **CICLO FOR**

***

Il ciclo for è un iteratore che esegue il blocco di istruzioni per un numero di volte specifico, a differenza del while che esegue il codice finché l'esito della condizione è False.

```python
# Esempio:
for i in range(10):
    print(i)
```

## **Funzione range()**

La funzione restituisce una sequenza di numeri, a partire da 0 per impostazione predefinita, incrementa di 1 (per impostazione predefinita) e si ferma al numero prima di quello che viene passato come parametro.

Tale funzione prende tre parametri:
1. start: numero di partenza dell'intervallo
2. stop: numero finale dell'intervallo (non compreso)
3. step: passo di avanzamento (non essenziale) 

```python
for i in range(1, 11, 2):
    print(i)
```

## **Nested loops** ##

Si possono utilizzare due loop ripetuti, uno all'interno dell'altro con una struttura chiamata **Nested Loop**. 
Dato un valore o variabile iniziale (esempio, una lista o una stringa) con _n_ numero di elementi e un secondo con _m_ numeri di elementi, un nested loop ripeterà la sua funzione _n x m_ volte, specificatamente ripetendo per ogni n del primo for loop, m volte il secondo loop. 


```python
#Esempio di Nested loop
lista1 = ["Ditto", "Charmender", "Pigeotto"]
lista2 = ["fa cagare", "è un pokemon", "sa fare la carbonara"]
for i in lista1:
    for j in lista2:
        print(i+" "+j)
```

## **Stringhe**

Una stringa è una sequenza immutabile di caratteri racchiusi tra singoli o doppi apici.

ES: `'Ciao'` oppure `"Ciao"`

Una stringa multiriga è invece racchiusa tra 3 apici singoli o doppi

ES: 
`""" Questa è una`
    
`stringa su più righe"""`

Per accedere ai caratteri di una stringa basta utilizzare le parentesi quadre indicando l'indice del carattere al quale si vuole accedere.

L'indice è un numero che indica la posizione di quel carattere all'interno della sequenza.

ES: `parola = "Ciao"`

`parola[0]` --> `'C'`

Il primo carattere di una stringa ha come indice il numero 0, mentre utilizzando un indice negativo si inizierà a contare dalla fine della stringa, quindi ad esempio: l'ultimo carattere di una stringa ha come indice il numero -1, oppure la lunghezza della stringa -1 (len(stringa)-1).

ES: `parola[-1]` --> `'o'`

È possibile anche selezionare una porzione di stringa, tale operazione viene detta **"slicing"** mentre la porzione di stringa viene detta **slice**. Per fare ciò basta utilizzare l'operatore `[n:m]` dove n è l'indice del carattere dal quale iniziare la selezione ed m è l'indice del carattere finale, carattere che non è compreso nella selezione.
Si può utilizzare l'operatore `[n:m:a]` il terzo carattere indica quali caratteri saltare all'interno della stringa. 

Se non viene indicato `n` la selezione inizierà dall'inizio della stringa, mentre se non viene indicato `m` allora la selezione sarà fino all'ultimo carattere della stringa, mentre se non vengono indicati `n` ed `m` allora verrà selezionata l'intera stringa.
Se si vogliono mantenere " o ' all'interno di una stringa, si considera che si possono utilizzare una o l'altra per indicare la stringa, e la seconda per mantenere il simbolo all'itnerno della stringa, oppure utilizzando \ prima del ". 
Ricorda = \n va accapo, \t lascia spazio. 
 

```python
# ES:
parola = "Ciao"
print(parola[:3])
print(parola[2:])
print(parola[1:3])
print(parola[:-1])
print(parola[-1:])
#skipping
strong= "GATTACA"
print(strong[1:5:2])
# ""
stringa = "cane 'infame'"
print(stringa)
stringa2 = 'cane\'infame'
print(stringa2)
```

## **Utilizzare i pattern**

Regular expression è una metodologia utilizzata per ricercare stringhe con alcune caratteristiche e non stringe specifiche. 
Si possono utilizzare i pattern per ricercare più di una stringa. Regular expression (regexps) è un formato di pattter per descrivere una serie di stringhe. 

**Regular** viene dal fatto che esse possono descrivere set infiniti che sono abbastanza regolari per avere una descrizione finita.
**Expression** viene dal fatto che possono essere costriuiti utilizzando operatori come espressioni matematiche.

Regular Expression sono utilizzate su python tramite la libreria **re**

```python
from re import *
text = "my name is Andrew"
result = search(r"[AT]",text)
print(result)
```

The **search** function takes a pattern and a string, and returns a (not easily
readable) **match object**
The r in r"[AT]" denotes it is a raw string, that is escape sequences are
dealt with as normal characters

```python
from re import *
text = "My name is Andrew"
result = search(r"[AT]",text)
print(result.start())
print(result.end())
print(text[result.start():result.end()])

```

## **ADDITIONS**
**Match** can be used as a Boolean operator

**[^P]** matches anything which is not a P

**.** matches any letter, if you require a .you need to put *\* before.

**[element]{numberofrepetitions}** repeat a specific element n times


The pattern [FILAPVM]{5} find matches in all of the
following:

AAAAA, 
AAPAP,
LAPMAVAILA,
VILLAMAP,
LAPLAMP,

And .{6} find matches in any string of at least length 6

```python
from re import *
pattern = r"[LIVMFE][FY]PWM[KRQTA]"
sequences = ["LHNEANLRIYPWMRSAGADR",
 "PTVGKQIFPWMKES",
 "NEANLKQIFPGAATR",
 "VFPWMKMGGAKGGESKRTR"]
for seq in sequences:
    match = search(pattern, seq)
    if match:
        print (seq, "matches")
    else:
        print (seq, "does not have the motif")
```

## **COUNT RANGES**

Sometimes you may want a range of repetitions,
For instance, a loop may have 3 to 5 residues in it.

**{m,n}** - repeat the previous pattern at least m times and
up to n times


For example, A{3,5} matches AAA, AAAA, and AAAAA
but does not match AA nor AATAA.

## **SPECIAL COUNT RANGES**
A few common count ranges have a special notation

**?** means 0 or 1 (equivalent to {0,1})

'*' means 0 or more

**+** means 1 or more

These are very common also outside bioinformatics and
regexps (e.g., the * for an unbounded number of
parameters in function definitions)



## **GROUPS**
We can use the parameter of start and end to select a specific
group

```python
from re import *
pattern = r"[ASD]{3,5}([LI])[^P]{2,5}"
seq = "EASALWTRD"
match = search(pattern, seq)
print(match)
print(match.start(),match.end())
print(match.start(1),match.end(1))
```

Groups are great for parsing. Suppose I have the string 

"Name: Andrew Age: 33"

We want to extract from the string the name and the age
values
We can use a pattern with a group for each field

"Name: ([^ ]+) +Age: ([0123456789]+)"

Starts with Name:
Then any sequence of at least one non space (group 1)

Then at least one space

Then Age:

Then any sequence of digits (group 2)

## **Some More shorthands**

There are shorthands for the most used sets

\d = [0123456789]

\w = letters, digits, and the underscore

\s = “whitespace” (space, newline, tab, and a few
others)

 For these it is important to use raw strings

```python
from re import *
text = "Name: Andrew Age: 33"
pattern = r"Name: ([^ ]+) +Age: (\d+)"
match = search(pattern, text)
print(match.start(1),match.end(1))
print(match.group(1),match.group(2))
```

**DEFINIRE UNA CLASSE**

**Le classi** sono blocchi che cominciano con una *keyword*, continuano con il nome della classe, definiscono una serie di funzioni, chiamate *metodi*.
Il primo parametro di ogni methodo è  **_self_**  che si riferisce all'oggetto della classe. Variabili che hanno come prefisso self sono variabili di istanza, una copia per ogni oggetto della classe. 

```python
class Counter:
    def __init__(self,value):
        self.val = value
    def tick(self):
        self.val = self.val + 1
    def reset(self):
        self.val = 0
    def value(self):
        return self.val
c1 = Counter(0)
c2 = Counter(3)
c1.reset()
c1.tick()
c2.tick()
print(c1.value())
print(c2.value())
```
